```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/drive/MyDrive/Kaggle/news-data.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data = data.drop_duplicates()
```

```python
data = data.reset_index()
```

```python
data.drop('index',inplace=True,axis=1)
```

```python
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

```python
corpus = []
for i in range(0, len(data)):
  text = re.sub('[^a-zA-Z]', ' ', data['text'][i])
  text = text.lower()
  text = ' '.join((word) for word in text.split() if word not in stopwords.words('english'))
  corpus.append(text)
```

```python
data['text'].sample(10)
```

```python
corpus[:3]
```

```python
from keras.preprocessing.text import Tokenizer
token = Tokenizer()
token.fit_on_texts(corpus)
```

```python
vocab_size = len(token.word_index) + 1
vocab_size
```

```python
print(token.index_word)
```

```python
encoded_text = token.texts_to_sequences(corpus)
```

```python
print(encoded_text[1:3])
```

```python
from keras.preprocessing.sequence import pad_sequences
maxlen = 500
X = pad_sequences(encoded_text,maxlen=maxlen, padding='post')
```

```python
len(X[0])
```

```python
len(X[1])
```

```python
X.shape
```

**Using GloVe**

```python
glove_vectors = dict()
```

```python
%%time
file = open('glove.6B.100d.txt', encoding='utf-8')

for line in file:
  values = line.split()
  word = values[0]
  vectors = np.asarray(values[1:])
  glove_vectors[word] = vectors
file.close()  
```

```python
len(glove_vectors.keys())
```

```python
glove_vectors.get('said').shape
```

```python
word_vector_matrix = np.zeros((vocab_size, 100))

for word, index in token.word_index.items():
  vector = glove_vectors.get(word)
  if vector is not None:
    word_vector_matrix[index] = vector
  else:
    print(word)  
```

```python
X
```

```python
word_vector_matrix.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense, Embedding, LSTM, SpatialDropout1D
from sklearn.model_selection import train_test_split
from keras.utils.np_utils import to_categorical
from keras.callbacks import EarlyStopping
from keras.layers import Dropout
```

```python
model = Sequential()
model.add(Embedding(vocab_size, output_dim=100, weights = [word_vector_matrix], input_length=maxlen, trainable=False))
model.add(LSTM(units=128))
model.add(Dense(5,activation='softmax'))
model.compile(optimizer='adam', loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

```python
Y = pd.get_dummies(data['category']).values
print('Shape of label tensor:', Y.shape)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.10, random_state = 42)
print(X_train.shape,Y_train.shape)
print(X_test.shape,Y_test.shape)
```

```python
model_history_1 = model.fit(X_train,Y_train,epochs=50,validation_split=0.3, batch_size=64)
```

```python
model_2 = Sequential()
model_2.add(Embedding(vocab_size, output_dim=100, weights = [word_vector_matrix], input_length=maxlen, trainable=False))
model_2.add(SpatialDropout1D(0.2))
model_2.add(LSTM(100, dropout=0.2, recurrent_dropout=0.2))
model_2.add(Dense(5, activation='softmax'))
model_2.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
print(model_2.summary())
```

```python
model_history = model_2.fit(X_train,Y_train,epochs=10,batch_size=64,validation_split=0.2)
```

```python
scores = model_2.evaluate(X_test, Y_test)
print(scores)
```
